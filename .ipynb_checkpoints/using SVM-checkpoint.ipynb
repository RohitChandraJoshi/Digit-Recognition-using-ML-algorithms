**Importing required dataset**

```python
y = train_data['label']
```

```python
X = train_data.drop(columns = 'label')
```

```python
print(train_data.shape)
```

```python
X = X/255.0
test_data = test_data/255.0

print("X:", X.shape)
print("test_data:", test_data.shape)
```

```python
X_scaled = scale(X)


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, train_size = 0.2 ,random_state = 10)
```

```python
kernels = ['linear','poly','sigmoid','rbf']
C = [1,10,20]
avg_scores = {}
for kval in kernels :
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel = kval, C = cval,gamma = 'auto'),X_train , y_train , cv =5 )
        avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)
        
avg_scores     
```

```python
model = SVC(C=20, gamma='auto', kernel="poly")

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
model.score(X_test , y_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.83)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_val_pca = pca.transform(X_test)

clf1 = SVC(kernel="poly",C=20,gamma='auto')
clf1.fit(X_train_pca,y_train)
predictions=clf1.predict(X_val_pca)
metrics.accuracy_score(y_test,predictions)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.83)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_val_pca = pca.transform(X_test)

clf1 = SVC(kernel="rbf",C=20,gamma='auto')
clf1.fit(X_train_pca,y_train)
predictions=clf1.predict(X_val_pca)
metrics.accuracy_score(y_test,predictions)
```
